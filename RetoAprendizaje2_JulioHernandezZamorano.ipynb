```python
# Importamos la biblioteca que nos sera util para medir el tiempo de ejecucion
import time

# Esta nos sera util para generar numeros aleatorios
import random

# Creamos una funcion unicamente para generar "n" cantidad de numeros flotantes
def generateRandomNumbers(n, onlyFloats):
  numbers = [] # Creamos una lista que irá almacenando elementos aleatorios
  for i in range(n): # Iteramos la candidad de veces del prametro usando un bucle
    if onlyFloats: #En caso de que solo queramos genear puros flotantes (CASO 1)
      numbers.append(random.uniform(0, 1000)) # Generamos un numero flotante aleatorio y lo agregamos a la lista
    else:  # Caso 2 y 3: Mezcla de enteros y flotantes
      if random.choice([True, False]):  # Decide si será un entero o un flotante
          numbers.append(random.randint(0, 1000))  # Genera un entero
      else:
          numbers.append(random.uniform(0, 1000))  # Genera un número decimal
  return numbers



# Creare una funcion cuyo unico trabajo sea retornarme el tiempo transcurrido
def getElapsedTime(startTime, endTime):
  return endTime - startTime # La diferencia nos permite saber el tiempo transcurrido

# Creare una funcion la cual reutilizare para imprimer resultados en ambos metodos
def printResults(startTime, endTime, sortedList, methodName):
    totalTime = getElapsedTime(startTime, endTime)

    #Esta forma de print con la f inicial nos permite insertar valores de variables
    print(f"Resultados del método {methodName}:") #Imprimimos nombre del metodo
    print(f"  Tiempo total: {totalTime}") #Imprimimos tiempo total calculdado
    print(f"  Lista ordenada: {sortedList}") # Imprimimos la lista ordenada
    print("-" * 100)  # Finalmente una línea separadora xD


# Creamos una funcion especifica para el metodo de ordenamiento burbuja
def bubbleSortMethodOptimized(list):
  sizeList = len(list) # Alamacenamos la longitud del lista recibida (input)
  # El metodo nos dice que tomemos el primer ciclo hasta longitud-1
  for i in range(sizeList - 1):
    #Esta variable hace mas eficiente el algoritmo cuando la lista ya se encuentra ordenada
    swapped = False #Agregue esta variable para detectar si hubo un intercambio
    # En ciclo anidado ajpra tenemos que restar 1
    for j in range(sizeList-i-1):
      # El elemento actual es mayor al elemento siguiente?
      if list[j] > list[j+1]:
        # Si es asi entonces se hace un intercambio de posicion
        list[j], list[j+1] = list[j+1], list[j]
        swapped = True
    if not swapped: #Si no hubo intercambios, entonces la lista ya esta ordenada
      break
  return list # Por ultimo regresamos la lista

# Creamos una funcion especifica para el metodo de ordenamiento burbuja
def bubbleSortMethod(list):
  sizeList = len(list) # Alamacenamos la longitud del lista recibida (input)
  # El metodo nos dice que tomemos el primer ciclo hasta longitud-1
  for i in range(sizeList - 1):
    # En ciclo anidado ajpra tenemos que restar 1
    for j in range(sizeList-i-1):
      # El elemento actual es mayor al elemento siguiente?
      if list[j] > list[j+1]:
        # Si es asi entonces se hace un intercambio de posicion
        list[j], list[j+1] = list[j+1], list[j]
      break
  return list # Por ultimo regresamos la lista


# Creamos una funcion especifica para el metodo quick sort
def quickSortMethod(list):
  # Si la lista tiene 0 o 1 elementos, ya está ordenada
  if len(list) < 2:
    return list  # Termina la ejecucion regresando la lista

  #En caso contrario, entra el metodo tenemos que evaluar
  else:
    pivot = random.choice(list)  # Elegir un pivote aleatorio de la lista
    left = [] # Crear las sublistas menores al pivote
    # Crear lista de iguales al pivote, esto mejora el rendimiento de quicksort
    # En especial cuando hay muchos repetidos
    iguales = []
    right = [] # Crear las sublistas mayores al pivote

    # Iteramos sobre cada elemento de la lista
    for currentElement in list:
      if currentElement < pivot: # El elemento actual es menor que el pivote?
        left.append(currentElement) # Entonces lo mandamos a la lista de menores
      elif currentElement == pivot: # Elemento actual es igual al pivote?
        iguales.append(currentElement) #Entonces lo mandamos a la lista de iguales
      else: #Solo entra si el elento actual es mayor
        right.append(currentElement) # Entonces lo mandaos a la lista de mayores

    # Combinar las sublistas ordenadas
    # Notese que iguales no tiene sentido procesar recursivamente
    # Esto es porque tiene el mismo valor que el pivote
    return quickSortMethod(left) + iguales + quickSortMethod(right)


# Funcion creada para procear un numero de iteraciones dado por parametro
def compareManyTimes(numberIterations, myList):
  WinQuickSort = 0 # Servira para contar cuantas veces fue mejor quicksort
  WinBubbleSort = 0 # Servira para contar cuantas veces fue mejor bubbleSort
  totalTimeQuickSort = 0 # Servira para contar el total de timpo en quicksort
  totaltimeBubbleSort = 0 # Servira para contar el total de tiempo en bubbleSort
  for i in range(numberIterations): # Creamos nuestro bucle segun el numero de iteraciones
    myListCopy = myList.copy() # Creamos una lista copia para procesar con quicksort

    startTimeBubbleSort = time.time() # Iniciamos a contar tiempo
    myList = bubbleSortMethod(myList) # Invocamos a la funcion de ordenamiento por burbuja
    endTimeBubbleSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento
    # Invocamos a nuestra funcion encargada de entregar el tiempo total transcurrido
    timeBubbleSort = getElapsedTime(startTimeBubbleSort, endTimeBubbleSort)

    startTimeQuickSort = time.time() # Iniciamos a contar tiempo
    myListCopy = quickSortMethod(myListCopy) # Invocamos a la funcion
    endTimeQuickSort = time.time() # Tomamos el tiempo una vez proceado el ordenamiento
    # Invocamos a nuestra funcion encargada de entregar el tiempo total transcurrido
    timeQuickSort = getElapsedTime(startTimeQuickSort, endTimeQuickSort)

    # Sumamos el tiempo total de ambos metodos
    totaltimeBubbleSort += timeBubbleSort
    totalTimeQuickSort += timeQuickSort

    if timeBubbleSort < timeQuickSort: # el tiempo de burbuja fue menor a quicksort?
      WinBubbleSort += 1 # Entonces burbuja fue mejor. Incrementamos 1
    else: # En caso contrario
      WinQuickSort += 1 # QuickSort Fue mejor. Incrementas su valor en 1

  # Imprimimos nuestras variables de mediciones para visualizar los resultados
  # Recordemos que la "f" al inicio del print sirve para ingresar variables de esta forma
  print(f"En {numberIterations} lista(s) de {len(myList)} elementos: ")
  print(f"BubbleSort fue mejor: {WinBubbleSort} de {numberIterations} veces")
  print(f"QuickSort fue mejor: {WinQuickSort} de {numberIterations} veces; Tiempo Promedio: {totalTimeQuickSort / numberIterations}")

  print() # Solo quiero dejar un espacio en blanco
  print(f"Tiempo Promedio con BubbleSort: {totaltimeBubbleSort / numberIterations}")
  print(f"Tiempo Total con BubbleSort: {totaltimeBubbleSort}")
  print() # Solo quiero dejar un espacio en blanco
  print(f"Tiempo Promedio con QuickSort {totalTimeQuickSort / numberIterations} ")
  print(f"Tiempo Total con QuickSort: {totalTimeQuickSort}")
  print() # Solo quiero dejar un espacio en blanco


# ======================== Datos Iniciales PARA CASO 1 =========================

# Lista de números tipo float de longitud igual a 500.
# Utilizaremos la funcion previamente creada para delegarle ese trabajo
# True es para indicar que solo queremos numeros flotantes
myList = generateRandomNumbers(500, True)
# Creamos una nueva lista copia para ser utilizada por el metodo quicksort
# Esto para tratar de ser justo y utilizar el mismo input
myListCopy = myList.copy()
print("***************** CASO: 1 (Lista de 500 flotantes) ********************")
print("Lista original ", myList) # Imprimimos la lista antes de ser ordenada
print() # Solo quiero dejar un espacio en blanco para visualizar mejor el resultado

# ****************** Proceso BubbleSort ******************************
startTimeBubbleSort = time.time() # Iniciamos a contar tiempo
myList = bubbleSortMethod(myList) # Invocamos a la funcion de ordenamiento por burbuja
endTimeBubbleSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento

# llamamos a nuestra funcion previamente creada para imprimir resultados
printResults(startTimeBubbleSort, endTimeBubbleSort, myList, "Burbuja")

# ****************** Proceso QuickSort ******************************
startTimeQuickSort = time.time() # Iniciamos a contar tiempo
# Utilizaremos la lista copia que creamos al inicio y NO esta ordenada
myListCopy = quickSortMethod(myListCopy) # Invocamos a la funcion de QuickSort
endTimeQuickSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento

# llamamos a nuestra funcion previamente creada para imprimir resultados
printResults(startTimeQuickSort, endTimeQuickSort, myListCopy, "QuickSort")





# ======================== Datos Iniciales PARA CASO 2 =========================

# Lista de números enteros y decimales de longitud igual a 1000
# Utilizaremos la funcion previamente creada para delegarle ese trabajo
# False es para indicar que NO solo queremos numeros flotantes
myList = generateRandomNumbers(1000, False)
# Creamos una nueva lista copia para ser utilizada por el metodo quicksort
# Esto para tratar de ser justo y utilizar el mismo input
myListCopy = myList.copy()
print("************ CASO: 2 (Lista de 1000 enteros y flotantes) **************")
print("Lista original ", myList)
print() # Solo quiero dejar un espacio en blanco para visualizar mejor el resultado

# ****************** Proceso BubbleSort ******************************
startTimeBubbleSort = time.time() # Iniciamos a contar tiempo
myList = bubbleSortMethod(myList) # Invocamos a la funcion de ordenamiento por burbuja
endTimeBubbleSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento

# llamamos a nuestra funcion previamente creada para imprimir resultados
printResults(startTimeBubbleSort, endTimeBubbleSort, myList, "Burbuja")

# ****************** Proceso QuickSort ******************************
startTimeQuickSort = time.time() # Iniciamos a contar tiempo
# Utilizaremos la lista copia que creamos al inicio y NO esta ordenada
myListCopy = quickSortMethod(myListCopy) # Invocamos a la funcion de QuickSort
endTimeQuickSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento

# llamamos a nuestra funcion previamente creada para imprimir resultados
printResults(startTimeQuickSort, endTimeQuickSort, myListCopy, "QuickSort")




# ======================== Datos Iniciales PARA CASO 3 =========================

# Lista de números enteros y decimales de longitud igual a 10000
# Utilizaremos la funcion previamente creada para delegarle ese trabajo
# False es para indicar que NO solo queremos numeros flotantes
myList = generateRandomNumbers(10000, False)
# Creamos una nueva lista copia para ser utilizada por el metodo quicksort
# Esto para tratar de ser justo y utilizar el mismo input
myListCopy = myList.copy()
print("************ CASO: 3 (Lista de 10000 enteros y flotantes) *************")
print("Lista original ", myList)
print() # Solo quiero dejar un espacio en blanco para visualizar mejor el resultado

# ****************** Proceso BubbleSort ******************************
startTimeBubbleSort = time.time() # Iniciamos a contar tiempo
myList = bubbleSortMethod(myList) # Invocamos a la funcion de ordenamiento por burbuja
endTimeBubbleSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento

# llamamos a nuestra funcion previamente creada para imprimir resultados
printResults(startTimeBubbleSort, endTimeBubbleSort, myList, "Burbuja")

# ****************** Proceso QuickSort ******************************
startTimeQuickSort = time.time() # Iniciamos a contar tiempo
# Utilizaremos la lista copia que creamos al inicio y NO esta ordenada
myListCopy = quickSortMethod(myListCopy) # Invocamos a la funcion de QuickSort
endTimeQuickSort = time.time() #Tomamos el tiempo una vez proceado el ordenamiento

# llamamos a nuestra funcion previamente creada para imprimir resultados
printResults(startTimeQuickSort, endTimeQuickSort, myListCopy, "QuickSort")







# Esto sucedera siempre?, hagamos varias pruebas con tamaños de lista diferentes
# Para esto encapsule el funcionamiento en la funcion descrita previamente
# En todos los metodos de esta prueba lo haremos iterando 100 veces
print("===================== Otras Comparaciones Interesantes =====================")
numberElementsInList = 10 # Primero con 10 elementos
myList = generateRandomNumbers(numberElementsInList, False)
compareManyTimes(100, myList) # Probaremos con 10 elementos

numberElementsInList = 500 # Ahora con 10 elementos
myList = generateRandomNumbers(numberElementsInList, False)
print("-" * 100)  # Otra línea separadora xD
compareManyTimes(100, myList) # Pobaremos con 500 elementos

numberElementsInList = 1000 # Ahora con 1000 elementos
myList = generateRandomNumbers(numberElementsInList, False)
print("-" * 100)  # Otra línea separadora xD
compareManyTimes(100, myList) # Probaremos ahora con 1000 elementos
print() # Solo quiero dejar un espacio en blanco


# Sera que los resultados cambien en los algoritmos si las listas ya estan ordenadas?
numberElements = 500
myList = list(range(1, numberElements + 1))  # Esto nos crea una lista ordenada de 1 hasta 1000
print(f"Lista Inicial Ordenada: {myList}")
myListCopy = myList.copy()
compareManyTimes(100, myList) # Iteraremos 100 veces una lista con 1000 elementos ya ordenados
print() # Solo quiero dejar un espacio en blanco

# Tiene ventaja alguno si la lista contiene muchos numeros repetidos
numberElements = 500
myList = [55] * numberElements  # Esto nos crea una de 1000 elementos todos con el valor 55
myListCopy = myList.copy()
print(f"Lista Inicial Repetida: {myList}")
myListCopy = myList.copy()
compareManyTimes(100, myList) # Iteraremos 100 veces una lista con 1000 elementos repetidos
```
